```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("C:\\Users\\mahek\\Downloads\\eda.csv")


selected_columns = ['koi_disposition', 'koi_period', 'koi_duration', 'koi_depth', 'koi_prad', 'koi_teq']


selected_df = df[selected_columns]

# Pairplot for selected features
sns.pairplot(selected_df, hue='koi_disposition')
plt.show()

# Correlation heatmap for selected features
correlation_matrix = selected_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()


```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='koi_disposition', data=df, palette='Set1')
plt.title('Countplot of koi_disposition')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x='koi_disposition', y='koi_teq', data=df, palette='Pastel1')
plt.title('Boxplot of koi_teq for Different Dispositions')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='koi_duration', y='koi_depth', hue='koi_disposition', data=df, palette='viridis')
plt.title('Scatter plot of koi_duration vs. koi_depth')
plt.xlabel('koi_duration')
plt.ylabel('koi_depth')
plt.show()

```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='ra', y='dec', hue='koi_disposition', data=df, palette='viridis', s=50)
plt.title('Scatter Plot of RA vs DEC with koi_disposition')
plt.xlabel('Right Ascension (RA)')
plt.ylabel('Declination (DEC)')
plt.legend(title='Disposition')
plt.show()
```
